```python
features_missing = data_with_missing_values[features_odi.columns]  # 특성 선택
```

```python
# 결측치를 채우는 함수를 정의합니다.
def fill_missing_values(row):
    if pd.isnull(row['ODI']):
        # 'ODI' 열이 결측치인 경우 모델을 사용하여 예측값 생성
        prediction = model.predict([row[features_odi.columns]])[0]
        return prediction
    else:
        # 결측치가 아닌 경우 현재 값 그대로 반환
        return row['ODI']
```

```python
# 결측치를 채워 넣습니다.
df_ROSL_finalset['ODI'] = df_ROSL_finalset.apply(fill_missing_values, axis=1)
```
