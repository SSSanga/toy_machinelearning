```python
# 결측치를 채우는 함수를 정의합니다.
def fill_missing_values(row):
    if pd.isnull(row['ODI']):
        # 'ODI' 열이 결측치인 경우 모델을 사용하여 예측값 생성
        prediction = model.predict([row[features_odi.columns]])[0]
        return prediction
    else:
        # 결측치가 아닌 경우 현재 값 그대로 반환
        return row['ODI']
```

```python
def fill_mean(column):
    mean = column.mean() if not column.isnull().all() else 0  # 열이 모두 null일 경우 0으로 채우도록 수정
    column.fillna(mean, inplace=True)
    return column
```

```python
mean_value = df_ROSL['통증기간(월)'].mean()
df_ROSL['통증기간(월)'].fillna(mean_value, inplace=True)
```
