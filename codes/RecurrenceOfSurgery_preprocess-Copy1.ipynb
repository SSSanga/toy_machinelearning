#### ODI를 def로 결측치 채우기 나머지는 대표값으로 넣어

```python
df_ROSL_drop = df_ROSL.dropna()
df_ROSL_drop
```

```python
target = df_ROSL_drop['ODI']
features = df_ROSL_drop[['Location of herniation','입원기간','통증기간(월)','수술시간','Seg Angle(raw)']]
target.shape, features.shape
```

```python
from sklearn.model_selection import train_test_split
features_train, features_test, target_train, target_test = train_test_split(features, target, random_state=111)
features_train.shape, features_test.shape, target_train.shape, target_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(features_train,target_train)
```

```python
model.coef_, model.intercept_
```

```python
df_ROSL_drop[200:250]
```

```python
model.predict(features_train[250:255])
```

```python
model.predict_proba(features_train[250:255])
```

```python
target_train_predict = model.predict(features_train)
target_train_predict.shape
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(target_train, target_train_predict)
```

```python
target_test_predict = model.predict(features_test)
target_test_predict.shape #target_test.shape과 동일
```

```python
accuracy_score(target_test, target_test_predict)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(target_train, target_train_predict))
```
